```python
import joblib
import gradio as gr

# Load the model
model = joblib.load(r"C:\Users\odne\Downloads\linear_regression_model.joblib")

# Define the prediction function
def predict(Month, DayOfWeek, CRSDepTime, CRSElapsedTime, AirTime, TaxiIn, TaxiOut, CarrierDelay, 
            WeatherDelay, NASDelay, SecurityDelay, LateAircraftDelay, Status, IsWeekend, 
            Fall, Spring, Summer, Winter):
    
    features = [Month, DayOfWeek, CRSDepTime, CRSElapsedTime, AirTime, TaxiIn, TaxiOut, CarrierDelay, 
                WeatherDelay, NASDelay, SecurityDelay, LateAircraftDelay, Status, IsWeekend, 
                Fall, Spring, Summer, Winter]
    
    prediction = model.predict([features])
    return prediction

# Create Gradio interface
interface = gr.Interface(
    fn=predict, 
    inputs=[gr.Number(label="Month"), 
            gr.Number(label="DayOfWeek"), 
            gr.Number(label="CRSDepTime"),
            gr.Number(label="CRSElapsedTime"),
            gr.Number(label="AirTime"),
            gr.Number(label="TaxiIn"),
            gr.Number(label="TaxiOut"),
            gr.Number(label="CarrierDelay"),
            gr.Number(label="WeatherDelay"),
            gr.Number(label="NASDelay"),
            gr.Number(label="SecurityDelay"),
            gr.Number(label="LateAircraftDelay"),
            gr.Number(label="Status"),
            gr.Checkbox(label="IsWeekend"),
            gr.Checkbox(label="Fall"),
            gr.Checkbox(label="Spring"),
            gr.Checkbox(label="Summer"),
            gr.Checkbox(label="Winter")],
    outputs="text"
)

# Launch the interface
interface.launch()
```
